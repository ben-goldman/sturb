```python
import numpy as np
import matplotlib.pyplot as plt
import h5py as h5
import tqdm
```

```python
s = h5.File("NS_isotropic_120_120_120.h5")
w = h5.File("NS_isotropic_120_120_120_w.h5")
```

```python
s["Turbulence"]["Ek"]["10"]
```

```python
plt.plot(np.array(s["Turbulence/Ek/1000"]))
plt.yscale("log")
plt.xscale("log")
```

```python
spec = np.ndarray((60, len(s["Turbulence/Ek"].keys())))
for i in range(0, 1000):
    t = str((i+1)*1)
    get = "Turbulence/Ek/" + t
    sp = np.array(s[get])
    spec[:, i] = sp
```

```python
spec
```

```python
plt.contourf(spec[:10, 500:520], norm="log")
# plt.yscale("log")
```

```python
for i in tqdm.tqdm(range(250)):
    t = str((i+1)*4)
    get = "P/3D/"+t
    p = w[get][:, :, 5]
    fig, ax = plt.subplots()
    a = ax.imshow(p)
    fig.colorbar(a)
    plt.savefig("frames/pic"+t+".png")
    plt.close()
```
