```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
f = h5py.File("MHD_w.h5")
ts = f["UB1/3D"].keys()
Es = np.ndarray((2, len(ts)))
for t in ts:
    print(t)
    U2 = f["UB0/3D/"+t][:, :, :]**2 + f["UB1/3D/"+t][:, :, :]**2 + f["UB2/3D/"+t][:, :, :]**2
    B2 = f["UB3/3D/"+t][:, :, :]**2 + f["UB4/3D/"+t][:, :, :]**2 + f["UB5/3D/"+t][:, :, :]**2
    U2 = np.mean(U2)
    B2 = np.mean(B2)
    Es[:, int(t)-1] = [U2, B2]
```
