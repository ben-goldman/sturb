```python
import numpy as np
import h5py as h5
import matplotlib.pyplot as plt
from scipy import stats
```

```python
pdfs = np.ndarray((100, 10))
for f in np.arange(1, 11):
    u = h5.File("snapshots/snapshots_s"+str(f)+".h5")["tasks"]["velocity_c"][:, :, :, :10]
    f_u2 = u[:, 0, 0, :,] #**2 + u[:, 1, 0, :]**2
    print(f_u2.shape)
    for t in range(10):
        pdfs[(f-1)*10 + t, :] = f_u2[t, :]
```

```python
plt.contourf(pdfs)
plt.savefig("hi.pdf")
```
